```python
daily_sales = supplement.groupby(['Date']).agg({
    'Revenue':'sum',
    'Units Sold':'sum',
    'Units Returned':'sum'
}).reset_index()
```

```python
supplement['returned_rate ']= supplement['Units Returned'] / supplement ['Units Sold'] 
supplement['returned_percentage ']=(( supplement['Units Returned'] / supplement ['Units Sold'] ) * 100 ).round(1)
```

```python
# supplement[['returned_rata ','Units Returned']]
# supplement.drop(['returned_rata'], inplace=True)
supplement[['returned_percentage ','Units Returned']].sort_values(by='returned_percentage ',ascending=False)
```
